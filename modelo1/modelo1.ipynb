```python
import pandas as pd
tabela = pd.read_csv("train.csv")
tabela2 = tabela.dropna(axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
codificador = LabelEncoder()

for coluna in tabela2.columns:
    if tabela2[coluna].dtype == "object" and coluna != "SalePrice":
        tabela2[coluna] = codificador.fit_transform(tabela2[coluna])
```

```python
x = tabela2.drop(["SalePrice", "Id", "MoSold", "YrSold", "Condition2"], axis=1)
y = tabela2["SalePrice"]

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25, random_state=42)

from sklearn.ensemble import RandomForestClassifier
tree_model = RandomForestClassifier()

tree_model.fit(x_treino, y_treino)

# Acurácia

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
valor_previsao = tree_model.predict(x_teste)
print("Modelo Random Forest:", mean_absolute_error(y_teste, valor_previsao))
print("Acurácia:", accuracy_score(y_teste, valor_previsao))
```
